Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score

import warnings
warnings.filterwarnings("ignore")
```

Data Collection & Processing

```python
df = pd.read_csv('/content/b_cancer_data.csv')
```

```python
df.head()
```

```python
# number of rows and columns in the dataset
df.shape
```

```python
# getting some information about the data
df.info()
```

```python
# checking for missing values
df.isnull().sum()
```

```python
# statistical measures about the data
df.describe([0.25,0.50,0.75,0.95])
```

0 --> Malignant

```python
df.groupby('label').mean()
```

Separating the features and target

```python
X = df.drop(columns='label', axis=1)
Y = df['label']
```

```python
X.head()
```

```python
Y.head(10)
```

Splitting the data into training data & Testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic Regression

```python
lgr = LogisticRegression()
```

```python
# training the Logistic Regression model using Training data

lgr.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
# accuracy on training data
X_train_prediction = lgr.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
```

```python
print('Accuracy on training data = ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = lgr.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
```

```python
print('Accuracy on test data = ', test_data_accuracy)
```

Random Forest

```python
rf = RandomForestClassifier()
```

```python
# training the Random Forest model using Training data

rf.fit(X_train, Y_train)
```

Accuracy  Score

```python
# accuracy on training data

X_train_prediction = rf.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
print('Accuracy on training data = ', training_data_accuracy)
```

```python
# accuracy on test data

X_test_prediction = lgr.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
print('Accuracy on test data = ', test_data_accuracy)
```

XGBoost

```python
xgb = XGBClassifier()
```

```python
# training the XGBoost model using Training data

xgb.fit(X_train, Y_train)
```

```python
# accuracy on training data

X_train_prediction = xgb.predict(X_train)
training_data_accuracy = accuracy_score(Y_train, X_train_prediction)
print('Accuracy on training data = ', training_data_accuracy)
```

```python
# accuracy on test data

X_test_prediction = xgb.predict(X_test)
test_data_accuracy = accuracy_score(Y_test, X_test_prediction)
print('Accuracy on test data = ', test_data_accuracy)
```

Hyperparameter Tuning

```python
#Tuning Hyperparameters for Random Forest Classifier

from sklearn.model_selection import GridSearchCV

# define models and parameters
rf_model = RandomForestClassifier()
n_estimators = [10, 100]
max_features=["sqrt", "log2", None]
min_samples_split=[2,3,4]


param_grid = {
    'n_estimators': n_estimators,
    'max_features': max_features,
    'min_samples_split' : min_samples_split
}


grid_search_forest = GridSearchCV(estimator=rf_model, param_grid=param_grid, n_jobs=-1,
                           scoring='r2',error_score=0,verbose=2,cv=2)

grid_search_forest.fit(X_train, Y_train)

# summarize results
print(f"Best: {grid_search_forest.best_score_:.3f} using {grid_search_forest.best_params_}")
```

Building a Predictive System

```python
def pred(vals,models):
  # change the input data to a numpy array
  input_data_as_numpy_array = np.asarray(vals)

  # reshape the numpy array as we are predicting for one datapoint
  input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

  for model in models:
    prediction = model.predict(input_data_reshaped)
    print(prediction)

    if (prediction[0] == 0):
      print(f'for {model} model: The Breast cancer is Malignant')

    else:
      print(f'for {model} model: The Breast Cancer is Benign')

input_data = (13.54,14.36,87.46,566.3,0.09779,0.08129,0.06664,0.04781,0.1885,0.05766,0.2699,0.7886,2.058,23.56,0.008462,0.0146,0.02387,0.01315,0.0198,0.0023,15.11,19.26,99.7,711.2,0.144,0.1773,0.239,0.1288,0.2977,0.07259)
models = [lgr, rf, xgb]
pred(input_data, models)
```
